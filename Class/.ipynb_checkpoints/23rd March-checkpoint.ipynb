# We are using Tensorflow with MNIST

Tensorflow makes statical graph with all the variables and placeholders, doesn't "run" the cells immediately during initialization

```python
num_pixels = 784
num_classes = 10
X = tf.placeholder(tf.float32, [None, num_pixels]) # Placeholder because we allocate ae space for the constant. None because arbitrary dimension. We can put in any batch size.
y_ = tf.placeholder(tf.float32, [None, num_classes])
```

```python
W1 = tf.Variable(tf.zeros([num_pixels, 256])) # Variable because this will change
W2 = tf.Variable(tf.zeros([256, 256]))
W3 = tf.Variable(tf.zeros([256, num_classes]))
```

```python
layer_1 = tf.nn.relu(tf.matmul(X, W1))
layer_2 = tf.nn.relu(tf.matmul(layer_1, W2))
y = tf.nn.softmax(tf.matmul(layer_2, W3))
```

```python
# y = tf.nn.softmax(tf.matmul(X, W))
```

```python
# In numpy, cross entropy loss:
# y1 = np.ones((10,)) <- real
# y2 = np.ones((10,)) * 0,1 <- predicted
# loss = np.mean(np.sum( - y1 * np.log(y2)))
```

```python
cost = - tf.reduce_mean( tf.reduce_sum( y_ * tf.log(y), axis=1))
```

```python
cost
```

```python
train_step = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    
    for i in range(2000):
        batch_xs, batch_ys = mnist.train.next_batch(100)
        _, loss = sess.run([train_step, cost], feed_dict= { X: batch_xs, y_ : batch_ys })
        print("Loss {} at iteration {}".format(loss, i+1))
```
